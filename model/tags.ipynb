```python
import os
```

```python
lines = []
extension = ".txt"
for item in os.listdir('../dataset'):
    if extension in item:
        with open('../dataset/' + item, 'r+') as F:
            lines.append(F.readlines())
lines
```

```python
path_prefix = '../dataset/'

with open('./dataset.txt', 'w') as F:
    for i in range(0, len(lines)):
        num = 0
        for line in lines[i]:
            # set maximum for training images 
            if num < 500:
                if line[0:2] == "./":
                    F.write(path_prefix + line[2: -1] + ' ' + str(i) + '\n')
                    # print line[2: -1]
                    # print line
                else:
                    pass
                    F.write(path_prefix + line[: -1] + ' ' + str(i) + '\n')
                    # print line[: -1]
                num += 1
```
